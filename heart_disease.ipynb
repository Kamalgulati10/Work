```python
from sklearn.model_selection import train_test_split
```

## Splitting the data

```python
X=heart_disease.drop("target",axis=1)
y=heart_disease["target"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

## Trying model

```python
np.random.seed=7
ran=RandomForestClassifier()
log=LogisticRegression(max_iter=1000)
```

```python
ran.fit(X_train,y_train)
```

```python
log.fit(X_train,y_train)
```

```python
ran.score(X_train,y_train)
```

```python
log.score(X_train,y_train)
```

```python
ran_pred=ran.predict(X_test)
```

```python
ran.score(X_test,y_test)
```

```python
log_pred=log.predict(X_test)
```

```python
log.score(X_test,y_test)
```

## Accuracy Test

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
```

```python
def accuracy_test(y_true,y_preds):
    acc=accuracy_score(y_true,y_preds)
    pre=precision_score(y_true,y_preds)
    rec=recall_score(y_true,y_preds)
    f1=f1_score(y_true,y_preds)
    
    metric_dict:{"acc":round(acc,2),
                "pre":round(pre,2),
                "rec":round(rec,2),
                "f1":round(f1,2),
                }
    print(f"accuracy={acc*100:.2f}")
    print(f"precision={pre*100:.2f}")
    print(f"recall={rec*100:.2f}")
    print(f"f1={f1*100:.2f}")
    
    
```

```python
accuracy_test(y_test,ran_pred)
```
